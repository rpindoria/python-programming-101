```
# definig functions for personal bot program
# by Ravji Pindoria 22 June 2020
# during programming 101, Introduction to python for educators - Futurelearn.com

def adding(input1, input2):
    print("Lets add some numbers")
    input1 = input("Number 1> ")
    input2 = input("Number 2> ")
    number1 = int(input1)
    number2 = int(input2)
    return number1 + number2


def multiply(input1, input2):
    print("Lets multiply some numbers")
    input1 = input("Number 1> ")
    input2 = input("Number 2> ")
    number1 = int(input1)
    number2 = int(input2)
    return number1 * number2


def difference(input1, input2):
    print("Lets subtract some numbers")
    input1 = input("Number 1> ")
    input2 = input("Number 2> ")
    number1 = int(input1)
    number2 = int(input2)
    return number1 - number2


def quotient(input1, input2):
    print("Lets take quotient of a numbers")
    input1 = input("Number 1> ")
    input2 = input("Number 2> ")
    number1 = int(input1)
    number2 = int(input2)
    return number1 / number2


def average():
    how_many = input("How many numbers> ")
    how_many = int(how_many)
    total = 0
    for number_count in range(how_many):
        number = input("Enter number " + str(number_count) + "> ")
        total = total + int(number)
    result = total / how_many
    print("the average = " + str(result))


def shopping_list():
    print("def for shopping list")

    shopping = []
    how_many = input("how many items of shopping do you have?")
    how_many = int(how_many)
    for item_number in range(how_many):
        item = input("what is item number" + str(item_number) + "?")
        shopping.append(item)
    print(shopping)

    how_many_required = input("how many items of shopping do you need?")
    how_many_required = int(how_many_required)

    itemsreq = how_many_required
    print("we need following items from our list: ")
    for i in range(itemsreq):
        print(shopping[i])
    cost = itemsreq * 5
    print("cost of ", itemsreq, "items =", "£", cost)


def resturantBill():
    number_of_diners = input("How many of you are at the meal? > ")
    how_much_tip = input("What percentage tip would you like to add to the bill ?>")
    what_is_the_bill = input("How much is the bill for? >")

    no_of_diners = int(number_of_diners)
    percentage = int(how_much_tip)
    meal_cost = int(what_is_the_bill)

    total_bill = meal_cost + (percentage * meal_cost / 100)
    each_person_owes = total_bill / no_of_diners
    print("Each person owes > £" + str(each_person_owes))


#############################################################################
##Functional calls for personal bot program.
finished = False
while finished == False:

    print("Hi, I am Marvin, your personal bot.")
    command = input("Select add or multiply or difference or quotient, or averages or shopping_list! or resturantBill")
    if command == "add":
    # calling function called adding with two parameters
        ans = adding(int, int)
        print("The sum is ", ans)

    elif command == "multiply":
    # calling function called multiply passing two parameters
        ans = multiply(int, int)
        print("Product is ", ans)

    elif command == "difference":
    # function call for subtracting - passing two parametersnum
        ans = difference(int, int)
        print("The Difference is ", ans)

    elif command == "quotient":
        ans = quotient(int, int)
    # function call for quotient - passing two paraeter
        print("The quotient is ", ans)

    elif command == "average":
        average()

    elif command == "shopping_list":
        shopping_list()

    elif command == "resturantBill":
        resturantBill()

    else:
        print("bye")
        command == "bye"
        finished = True
```
